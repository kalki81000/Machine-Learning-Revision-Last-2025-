```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#normal distribution.gaussian>> majority of data point is clsutered around mean

mu = 0
sigma = 0.1

s = np.random.normal(mu, sigma, size = 10000)
```

```python
len(s)
```

```python
plt.hist(s, 30)
plt.show()
```

```python
sns.distplot(s)
```

```python
#uniform distribution >> all the outcomes has equal probability to occur in a range
u = np.random.uniform(0, 1, 10000)
plt.hist(u)
plt.show()
```

```python
sns.distplot(u)
```

```python
#log normal distribution
l = np.random.lognormal(mean = 0, sigma = 0.5, size = 1000)
l = l[np.where(abs(l) < 5)]
plt.hist(l, bins =100)
plt.show()
```

```python
sns.boxplot(l)
```

```python
sns.distplot(l)
```

```python
#lognormal >> normal distr >> log normal dist

sns.distplot(np.log(l))
```

```python
#square root transformation
sns.distplot(np.sqrt(l))
```

```python
#chisquare distribution
c = np.random.chisquare(df = 30, size = 10000)

c = c[np.where(abs(c) < 100)]

plt.hist(c, bins = 1000)

plt.show()
```

```python
#multiple chisquare dist in one plot
c = np.random.chisquare(df = 30, size = 10000)
d = np.random.chisquare(df = 1, size = 10000)
c = c[np.where(abs(c) < 100)]
d = d[np.where(abs(d) < 100)]
plt.hist(c, bins = 1000)
plt.hist(d, bins = 1000)
plt.show()
```

```python
#t distribution >>> sample size < 30

t = np.random.standard_t(df = 1, size = 10000)
t = t[np.where(abs(t) < 10)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
t = np.random.standard_t(df = 10, size = 10000)
t = t[np.where(abs(t) < 10)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
#F distribution
f = np.random.f(dfnum = 100, dfden = 100, size=10000)
f = f[np.where(abs(f) < 10)]
plt.hist(f, bins = 1000)
plt.show()
```

```python
#discrete distribution
#bernoulli dist with 1 trial is called bernoulli

b = np.random.binomial(n = 1, p = 0.5, size = 1000)
plt.hist(b)
plt.show()
```

```python
b = np.random.binomial(n = 100, p = 0.5, size = 1000)
plt.hist(b)
plt.show()
```

```python
#poisson distribution
p = np.random.poisson(lam = 5, size = 1000)
plt.hist(p, bins = 20)
plt.show()
```
