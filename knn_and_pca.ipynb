```python
object_columns = df.dtypes[df.dtypes == "object"].index
```

```python
for col in object_columns:
    print(col, ">>>", df[col].unique())
```

```python
df_dummy = pd.get_dummies(df[object_columns], drop_first=True, dtype = int)
```

```python
df.drop(object_columns, inplace = True, axis=1)
```

```python
df_ = pd.concat([df, df_dummy], axis = 1)
```

```python
df_.dtypes
```

```python
df_
```

```python
X = df_.drop("revenue", axis = 1)
y = df_['revenue']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_train = pca.fit_transform(X_train)
```

```python
pca.explained_variance_ratio_
```

```python
X_train.shape
```

```python
X_test = pca.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
reg = KNeighborsRegressor()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors':[3, 5, 7, 9, 11, 13, 21],
    'algorithm':['auto', 'ball_tree', 'kd_tree']
}
```

```python
grid = GridSearchCV(estimator=reg, param_grid=param_grid, cv=5, verbose=3)
```

```python
grid
```

```python
grid.fit(X, y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
